## Notes:
So I did struggle initially to make the fonts work on my installation on **Ubuntu 20.04** as some fonts were not available such as `Comic Sans MS` which was a font I saw being used in many online tutorials (just for examples obviously). 

After some digging around I found that you must check what fonts are available on your system before trying any font to avoid errors. I found a [website](http://jonathansoma.com/lede/data-studio/matplotlib/list-all-fonts-available-in-matplotlib-plus-samples/) that provided the below code cell for figuring this out. 
It essentially prints out the name of the font as well as how it looks.

This is definitly a code cell worth keeping in the notebook and for reference in the future.

```python
import matplotlib.font_manager
from IPython.core.display import HTML

def make_html(fontname):
    return "<p>{font}: <span style='font-family:{font}; font-size: 24px;'>{font}</p>".format(font=fontname)

code = "\n".join([make_html(font) for font in sorted(set([f.name for f in matplotlib.font_manager.fontManager.ttflist]))])

HTML("<div style='column-count: 2;'>{}</div>".format(code))
```

## The tick marks need not be evenly spaced on the graph
Also the tick marks need not be passed in any particular order to the functions as well, it is just a convention to pass the values in ascending order

```python
plt.plot(x, y)

plt.title("The Graph", fontdict=FONT_DICT_TITLE)
plt.xlabel("X Axis", fontdict=FONT_DICT_X_AXIS)
plt.ylabel("Y Axis", fontdict=FONT_DICT_Y_AXIS)

# The ticks automatically resize the graph
plt.xticks([i for i in np.arange(0, 10, 0.5)])  # The values need not be evenly spaced
plt.yticks([i for i in np.arange(1, 100, 10)] + [5, 95, 100])  # The values need not be evenly spaced

plt.ylim(0)  # Makes sure that the 0 point on the Y-Axis is at the bottom left corner of the graph only

plt.show()  # This is actually the right way to show the graph
```

## Shorthand notation for the color marker and linestyle
### [Click here](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot) and see the notes section for the different formats for the format string

```python
# Use shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x, y, "mo--", label="x^2",
         markersize=7, markeredgecolor='black', linewidth=3)

plt.title("The Graph", fontdict=FONT_DICT_TITLE)
plt.xlabel("X Axis", fontdict=FONT_DICT_X_AXIS)
plt.ylabel("Y Axis", fontdict=FONT_DICT_Y_AXIS)

# The ticks automatically resize the graph
plt.xticks([i for i in np.arange(0, 10, 0.5)])  # The values need not be evenly spaced
plt.yticks([i for i in np.arange(0, 100, 10)])  # The values need not be evenly spaced

plt.ylim(0)  # Makes sure that the 0 point on the Y-Axis is at the bottom left corner of the graph only
plt.legend()  # This needs the label kwarg to be defined during the plot stage

plt.show()  # This is actually the right way to show the graph
```

## Making 2 Graphs Look as One

```python
x = np.arange(0, 11, 1)
y = x ** 2
```

```python
# Use shorthand notation
# fmt = '[color][marker][line]'

# The plot values for the 2 graphs need to overlap by
# atleast by one value/point so that the graph looks continuous
plt.plot(x[:5], y[:5], "mo-", label="x^2",
         markersize=7, markeredgecolor='black', linewidth=3)  # 1st half of the graph
plt.plot(x[4:], y[4:], "mo--", label="x^2",
         markersize=7, markeredgecolor='black', linewidth=3)  # 2nd half of the graph


plt.title("The Graph", fontdict=FONT_DICT_TITLE)
plt.xlabel("X Axis", fontdict=FONT_DICT_X_AXIS)
plt.ylabel("Y Axis", fontdict=FONT_DICT_Y_AXIS)

# The ticks automatically resize the graph
plt.xticks([i for i in np.arange(0, 10, 0.5)])  # The values need not be evenly spaced
plt.yticks([i for i in np.arange(0, 101, 10)])  # The values need not be evenly spaced

plt.ylim(0)  # Makes sure that the 0 point on the Y-Axis is at the bottom left corner of the graph only
plt.legend()  # This needs the label kwarg to be defined during the plot stage

plt.show()  # This is actually the right way to show the graph
```

## Increasing the Figure Size

```python
# Use shorthand notation
# fmt = '[color][marker][line]'

# Create a graph of 1500x750 (width(X) x height(Y)) pixels image of the graph
plt.figure(figsize=(10, 5), dpi=150)  # figsize=(x_ratio, y_ratio)

# The plot values for the 2 graphs need to overlap by
# atleast by one value so that the graph looks continuous
plt.plot(x[:5], y[:5], "mo-", label="x^2",
         markersize=7, markeredgecolor='black', linewidth=3)  # 1st half of the graph
plt.plot(x[4:], y[4:], "m.--", label="x^2",
         markersize=6, markeredgecolor='black', linewidth=2)  # 2nd half of the graph


plt.title("The Graph", fontdict=FONT_DICT_TITLE)
plt.xlabel("X Axis", fontdict=FONT_DICT_X_AXIS)
plt.ylabel("Y Axis", fontdict=FONT_DICT_Y_AXIS)

# The ticks automatically resize the graph
plt.xticks([i for i in np.arange(0, 10, 0.5)])  # The values need not be evenly spaced
plt.yticks([i for i in np.arange(0, 101, 10)])  # The values need not be evenly spaced

plt.ylim(0)  # Makes sure that the 0 point on the Y-Axis is at the bottom left corner of the graph only
plt.legend()  # This needs the label kwarg to be defined during the plot stage

plt.savefig(os.path.join(".", "graph_images", "myFirstGraph.png"))  # Saves the graph in the current size and resolution

plt.show()  # This is actually the right way to show the graph
```

## Summary of Line Graphs

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(8,5), dpi=100)  # This property must be specified for every graph you make individually

# Line 1

# Keyword Argument Notation
#plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

# Shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'b^--', label='2x')

## Line 2

# select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'rx-', label='X^2')

# Plot remainder of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'rx--')

# Add a title (specify font parameters with fontdict)
plt.title('Our First Graph!', fontdict={'fontname': 'Khmer OS', 'fontsize': 20})

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# X, Y axis Tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4,])
#plt.yticks([0,2,4,6,8,10])

# Add a legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig(os.path.join(".", "graph_images", "myFirstGraph_summary.png"), dpi=300)

# Show plot
plt.show()
```
